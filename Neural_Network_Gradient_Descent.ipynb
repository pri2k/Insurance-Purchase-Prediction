```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('insurance_data.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age','affordibility']],df.bought_insurance,test_size=0.2, random_state=25)
```

```python
X_train_scaled = X_train.copy()
X_train_scaled['age'] = X_train_scaled['age'] / 100

X_test_scaled = X_test.copy()
X_test_scaled['age'] = X_test_scaled['age'] / 100
```

### **Model**

```python
model = keras.Sequential([
    keras.layers.Dense(1,input_shape = (2,) , activation = 'sigmoid', kernel_initializer = 'ones', bias_initializer = 'zeros')
])

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train_scaled, y_train, epochs = 20000)
```

### **Evaluate**

```python
model.evaluate(X_test_scaled,y_test)
```

```python
model.predict(X_test_scaled)
```

### **Functions**

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

print(sigmoid(1))
print(sigmoid(0))
print(sigmoid(-1))
print(sigmoid(10))
print(sigmoid(100))
print(sigmoid(-100))
print(sigmoid(50))
```

```python
def predict_function(age, affordibility, w1, w2, b):
  weight = w1*age + w2*affordibility + b
  return sigmoid(weight)
```

```python
def log_loss(y_true, y_predicted):
  epsilon = 1e-25
  y_predicted_new = np.clip(y_predicted,epsilon,1-epsilon)
  return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def gradient_descent(age, affordability, y_train, epochs):
  w1 = w2 = 1
  b = 0
  rate = 0.001
  n = len(age)
  for i in range(epochs):
    weight = w1*age + w2*affordability + b
    y_preds = sigmoid(weight)
    loss = log_loss(y_train, y_preds)
    w1_d = (1/n)* np.dot(np.transpose(age),y_preds-y_train)
    w2_d = (1/n)* np.dot(np.transpose(affordability),y_preds-y_train)
    w1 = w1 - rate*w1_d
    w2 = w2 - rate*w2_d
    b = b - rate*np.mean(y_preds-y_train)
    if i%1000 == 0:
      print(f'Epoch:{i}, w1:{w1}, w2:{w2}, b:{b}, loss:{loss}')
  return w1, w2, b

w1, w2, b = gradient_descent(X_train_scaled['age'],X_train_scaled['affordibility'],y_train,20000)
```

```python
print("my models weights", w1, w2, b)
print(model.get_weights())

print()
print("My Predictions")
print(predict_function(X_test_scaled['age'],X_test_scaled['affordibility'],w1,w2,b))
print()
print("Actual Predictions")
print(model.predict(X_test_scaled))
print()
print(y_test)
```
